```python
import pandas as pd
import numpy as np
import math, os
import matplotlib.pyplot as plt
```

#  Linear Regression Calculation Section

```python
# to calculate the least squares, get the:
# sum of all x's
# sum of all y's
# sum of all x^2's 
# sum of all xy's 
#calculate the slope 
#(N * (change in xy) - (change of x * change of y)) / (N * (change in x^2) - (change in x)^2)
#calculate intercept
#((change in y -  m * change in x) / N)

def leastSquaresRegression(csv, firstRow, secondRow):
    #initialize starting values
    x, y, xy, xx = 0, 0, 0, 0
    n = csv.shape[0]
    for index, item in csv.iterrows():
        #print('The item of {} was {} and the item of {}  was {}'.format(firstRow, item[firstRow], secondRow, item[secondRow]))
        getX = item[firstRow]
        getY = item[secondRow]
        if(np.isnan(item[firstRow]) or np.isnan(item[secondRow])):
            n -= 1
            continue
        x += getX
        y += getY
        xy += (getX * getY)
        xx += (getX * getX)
    #calculate the slope 
    #(N * (change in xy) - (change of x * change of y)) / (N * (change in x^2) - (change in x)^2)
    change = x * y
    top = xy*n
    top = top - change
    
    squareX = math.pow(x, 2)
    bottom = n*xx
    bottom = bottom - squareX
    
    m = top/bottom
    
    #calculate intercept
    #((change in y -  m * change in x) / N)
    right = m * x
    top = y - right
    
    bottom = n
    
    b = top/bottom
    
    #print('The best fit for the {} and {} graph would be: {}x + {}'.format(firstRow,  secondRow, m, b))

    return([m,b])
```

```python
# Calculate the confidence of our 'best fit line'
# To do this, we are going to calcualte the R-squared coefficient of determination
# Starting, we need to caculate the Square Error of said line 
# (y1 - (mx1 + b))^2 + (y1 - (mx1 + b))^2 + .. + (y1 - (mx1 + b))^2
# Then, we get the total variation of y 
# (y1 -  mean of y)^2 + (y2 -  mean of y)^2 + ... +(yn -  mean of y)^2
# finally, calculate the coefficient of determination 
# 1 - ((Square Error) / ( Total Determination of y))
# Values closer to 1 mean the line is a good fit. Values closer to 0 mean the line is a bad fit

#first row should be x, secondRow is y
def linRegConfidence(csv, firstRow, secondRow, m, b):
    meanY = csv[secondRow].mean()
    squareError = 0
    totalDeterminate = 0
    for index, item in csv.iterrows():
        if(np.isnan(item[firstRow]) or np.isnan(item[secondRow])):
            continue
        #grab essential numbers for the calculations
        y = item[secondRow]
        x = item[firstRow]
        
        #calculate the Square Error
        predictedY = (m * x) + b
        inside = y - predictedY
        SE = math.pow(inside, 2)
        squareError += SE
        
        #calculate the total determination of y
        inside = y - meanY
        TDY = math.pow(inside, 2)
        totalDeterminate += TDY
        
    #calculate the coefficeint of determination
    coefficient = 1 - (squareError/totalDeterminate)
   # print('The Square Error was: {}\nThe Total Determinate of Y was: {}\nMaking the Coefficeint of determination: {}'
          #.format(squareError, totalDeterminate, coefficient))
    return(coefficient)
```

# Quadratic Regression Calculation Section

```python
def multiplyArr(arr, val):
    for i in range(0, len(arr)):
        arr[i] = arr[i] * val
        #print(arr[i])
    return(arr)

#add all values from the first array, to the second
def addArr(first, second):
    newArr = []
    for i in range(0, len(first)):
        newArr.append(first[i] + second[i])
    return(newArr)

# This function solves for a,b,c that was generated in the leastSquareQuadratic funciton.
# Three equations, for example:
    # First Equation: 9669.0a + 1225.0b + 165.0x = 5174.3
    # Second Equation: 1225.0a + 165.0b + 25.0x = 809.6999999999999
    # Third Equation: 165.0a + 25.0b + 5x = 167.1

def solveForVariables(arr):
    newArr = arr
    #print('In Element WISE')
    
    # Find the value needed to eliminate the first equation b with the second equation b. 
    # Then, multiply with all values in the array.
    # Add new values of first equation to the second. We now only have 2 variables to deal with
    remove = abs(arr[0][1])/abs(arr[1][1])
    if((arr[1][1] > 0 and arr[0][1] > 0) or (arr[1][1] < 0 and arr[0][1] < 0)):
        remove = remove * -1
        newArr[1] = multiplyArr(newArr[1], remove)
    else:
        newArr[1] = multiplyArr(newArr[1], remove)
        
    #a and c are now isolated in the first equation
    firstEQ = addArr(newArr[1], arr[0])
    
    # Find the second equation that removes b.
    remove = abs(arr[2][1])/abs(arr[0][1])
    if((arr[2][1] > 0 and arr[0][1] > 0) or (arr[2][1] < 0 and arr[0][1] > 0)):
        remove = remove * -1
        newArr[0] = multiplyArr(newArr[0], remove)
    else:
        newArr[0] = multiplyArr(newArr[0], remove)
    #a and c are now isolated in the second equation
    secondEQ = addArr(newArr[0], arr[2])
    #print(firstEQ)
    #print(secondEQ)
    
    # Find a:
    #use firstEQ and secondEQ to solve for a
    remove = abs(firstEQ[2])/abs(secondEQ[2])
    if( (secondEQ[2] > 0 and firstEQ[2] > 0)  or (secondEQ[2] < 0 and firstEQ[2] > 0)):
        remove = remove * -1
        newArr = multiplyArr(secondEQ, remove)
    else:
        newArr = multiplyArr(secondEQ, remove)
    aArr = addArr(secondEQ, firstEQ)
    # solve for a
    a = aArr[3]/aArr[0]
    
    #plug a into firstEQ to solve for c
    firstEQ[0] = firstEQ[0] * a
    firstEQ[3] = firstEQ[3] - firstEQ[0]
    c = firstEQ[3] / firstEQ[2]
    
    #plug a and c into arr[2] to now solve for b
    solveB = arr[2]
    solveB[0] = solveB[0] * a
    solveB[2] = solveB[2] * c
    solveB[3] = solveB[3] - solveB[0]
    solveB[3] = solveB[3] - solveB[2]
    solveB[3]  = solveB[3]/solveB[1]
    b = solveB[3]

    #print('a was: {} b was: {} c was: {}'.format(a,b,c))
    return([a,b,c])
```

```python
# for Least Square Quadratic calculations, we need X^4, X^3, X^2, X, Y, X*Y, X^2*Y
# generate a matrix:
    # sum(x^4)+sum(x^3)+sum(x^2)=sum(x^2*y)
    # sum(x^3)+sum(x^2)+sum(x)=sum(x*y)
    # sum(x^2)+sum(x)+sum(number of rows) = sum(y)
# solve matrix for a, b, and c
# quadratic equation = a*x^2 + b*x + c

def leastSquareQuadratic(csv, firstRow, secondRow):
    x4, x3, x2, x, y, x2y, xy = 0,0,0,0,0,0,0
    n = csv.shape[0]
    for index, item in csv.iterrows():
        # confirm we have data to use for the current row
        if(np.isnan(item[firstRow]) or np.isnan(item[secondRow])):
            n -= 1
            continue
        #grab essential values
        getx = item[firstRow]
        gety = item[secondRow]
        place_x2 = math.pow(getx, 2)
        place_xy = getx * gety
        place_x2y = place_x2 * gety
        #calculate sums we need for least square quadratic
        x4 += math.pow(getx, 4)
        x3 += math.pow(getx, 3)
        x2 += place_x2
        x  += item[firstRow]
        y  += item[secondRow]
        xy += place_xy
        x2y += place_x2y
        
    arr = [
        [x4,x3,x2,x2y], 
        [x3,x2,x,xy],
        [x2,x,n,y]
    ]
    #print('First Equation: {}a + {}b + {}x = {}'.format(x4,x3,x2,x2y))
    #print('Second Equation: {}a + {}b + {}x = {}'.format(x3,x2,x,xy))
    #print('Third Equation: {}a + {}b + {}x = {}'.format(x2,x,n,y))
    return(solveForVariables(arr))
```

```python
def quadRegConfidence(csv, firstRow, secondRow, a, b, c):
    meanY = csv[secondRow].mean()
    squareError = 0
    totalDeterminate = 0
    for index, item in csv.iterrows():
        if(np.isnan(item[firstRow]) or np.isnan(item[secondRow])):
            continue
        #grab essential numbers for the calculations
        y = item[secondRow]
        x = item[firstRow]
        
        #calculate the Square Error
        predictedY = (x ** 2 * a) + (b * x) + c
        inside = y - predictedY
        SE = math.pow(inside, 2)
        squareError += SE
        
        #calculate the total determination of y
        inside = y - meanY
        TDY = math.pow(inside, 2)
        totalDeterminate += TDY
        
    #calculate the coefficeint of determination
    coefficient = 1 - (squareError/totalDeterminate)
    #print('The Square Error was: {}\nThe Total Determinate of Y was: {}\nMaking the Coefficeint of determination: {}'
          #.format(squareError, totalDeterminate, coefficient))
    return(coefficient)
```

# Train and Save Predicted Data

```python
def train():
    # Delete CSV for trained data if the file already exists.
    if(os.path.exists('./equations.csv')):
        os.remove('./equations.csv')
    
    # Training csv
    csv = pd.read_csv('./train.csv')

    # Grab the different Column names we are going to use for comparison
    vals = csv.columns.values

    for val in vals:
        if(csv[val].dtype == np.dtype('int64') or csv[val].dtype == float):
            # Find the linear equation for the current column
            l = leastSquaresRegression(csv, val, 'SalePrice')
            m = l[0]
            i = l[1]
            
            # Get the confidence of our best fit line
            linConf = linRegConfidence(csv, val, 'SalePrice', m, i)
            
            # Find the quadratic equation for the current column
            q = leastSquareQuadratic(csv, val, 'SalePrice')
            a = q[0]
            b = q[1]
            c = q[2]
            
            # Get the confidence of our best fit line
            quadConf = quadRegConfidence(csv, val, 'SalePrice', a, b, c)
            
            if(linConf < .3 and quadConf < .3 or linConf == 1 or quadConf == 1):
                continue
            
            # Add the trained data to CSV file 
            if(linConf > quadConf):
                data = {
                    'Column': val,
                    'Type': 'Linear',
                    'm': m,
                    'i': b,
                    'a': [0],
                    'b': [0],
                    'c': [0]
                }
                createCSV(data)
            else:
                data = {
                    'Column': val,
                    'Type': 'Quadratic',
                    'm': [0],
                    'i': [0],
                    'a': a,
                    'b': b,
                    'c': c
                }
                createCSV(data)
```

```python
def createCSV(data):
    
    df = pd.DataFrame(data)
    
    # File already exists, do not add headers
    if(os.path.exists('./equations.csv')):
        df.to_csv('equations.csv', mode='a', index=False, header=False)
    # File does not exist, add headers
    else:
        df.to_csv('equations.csv',index=False)
```
